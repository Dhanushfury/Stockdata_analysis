```python
import os
import yaml
import csv


input_dir = r"C:\Users\Nivetha s\Desktop\prject 2\2023-10"
output_dir = r"C:\Users\Nivetha s\Desktop\prject 2\csv output"


os.makedirs(output_dir, exist_ok=True)


for filename in os.listdir(input_dir):
    if filename.endswith(".yaml") or filename.endswith(".yml"):
        yaml_path = os.path.join(input_dir, filename)
        csv_path = os.path.join(output_dir, filename.replace(".yaml", ".csv").replace(".yml", ".csv"))

        with open(yaml_path, "r", encoding="utf-8") as yaml_file:
            data = yaml.safe_load(yaml_file)

        
        if isinstance(data, dict):
            data = [data]

       
        keys = set()
        for item in data:
            keys.update(item.keys())

       
        with open(csv_path, "w", newline="", encoding="utf-8") as csv_file:
            writer = csv.DictWriter(csv_file, fieldnames=list(keys))
            writer.writeheader()
            writer.writerows(data)

        print(f"Converted {filename} to {csv_path}")

print("Conversion completed!")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Path to folder with 50 stock CSV files
folder_path =(r"C:\Users\Nivetha s\Desktop\prject 2\50 stocks")
volatility_data = []

# Loop through each CSV stock file
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        filepath = os.path.join(folder_path, filename)
        try:
            df = pd.read_csv(filepath)

            if 'close' in df.columns:
                df = df.dropna(subset=['close'])  # Remove NaN closes
                df['returns'] = df['close'].pct_change()
                volatility = df['returns'].std()

                stock_name = filename.replace(".csv", "")
                volatility_data.append({'Stock': stock_name, 'Volatility': volatility})
        except Exception as e:
            print(f" Error reading {filename}: {e}")

# Convert to DataFrame
volatility_df = pd.DataFrame(volatility_data)

# Save all 50 stocks' volatility to CSV
output_csv = os.path.join(folder_path, 'volatility_50.csv')
volatility_df.to_csv(output_csv, index=False)
print(f" Saved all volatility data to: {output_csv}")

# Sort and get top 10 most volatile stocks
top10_df = volatility_df.sort_values(by='Volatility', ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(top10_df['Stock'], top10_df['Volatility'], color='orange')
plt.title("Top 10 Most Volatile Stocks (Past Year)")
plt.ylabel("Volatility (Standard Deviation of Daily Returns)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Folder with stock files
folder_path = r"C:\Users\Nivetha s\Desktop\prject 2\50 stocks"
cumulative_list = []

# Read and compute cumulative returns
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        try:
            filepath = os.path.join(folder_path, filename)
            df = pd.read_csv(filepath)

            if 'date' in df.columns and 'close' in df.columns:
                df = df.dropna(subset=['close'])

                # Convert to datetime and remove invalid dates
                df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
                df = df.dropna(subset=['date'])

                # Sort and remove duplicates
                df = df.sort_values('date')
                df = df.drop_duplicates(subset='date', keep='first')

                # Calculate returns
                df['daily_return'] = df['close'].pct_change()
                df['cumulative_return'] = (1 + df['daily_return']).cumprod()
                df['Stock'] = filename.replace(".csv", "")
                
                cumulative_list.append(df[['date', 'Stock', 'cumulative_return']])
        except Exception as e:
            print(f" Error reading {filename}: {e}")

# Combine all into a single DataFrame
combined_df = pd.concat(cumulative_list)
combined_df.to_csv(os.path.join(folder_path, "cumulative_returns.csv"), index=False)
print(" Saved cumulative return data to: cumulative_returns.csv")

# Get final cumulative return for each stock
latest_returns = combined_df.sort_values('date').groupby('Stock').tail(1)
top5_stocks = latest_returns.sort_values('cumulative_return', ascending=False).head(5)['Stock'].tolist()

# Plot cumulative return for top 5 stocks
plt.figure(figsize=(12, 6))
for stock in top5_stocks:
    stock_df = combined_df[combined_df['Stock'] == stock]
    plt.plot(stock_df['date'], stock_df['cumulative_return'], label=stock)

plt.title("Cumulative Return Over Time  Top 5 Performing Stocks")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Step 1: Load Sector Mapping and Extract Tickers ===
sector_path = r"C:\Users\Nivetha s\Desktop\prject 2\3. Sector-wise Performance\Sector_data - Sheet1.csv"
sector_df = pd.read_csv(sector_path)

# Extract ticker from 'Symbol' column after colon
sector_df['Ticker'] = sector_df['Symbol'].str.split(":").str[-1].str.strip()

# === Step 2: Load Cumulative Returns Data ===
returns_path = r"C:\Users\Nivetha s\Desktop\prject 2\2. Cumulative Return Over Time\cumulative_returns.csv"
returns_df = pd.read_csv(returns_path)

# === Step 3: Get Latest Cumulative Return Per Stock ===
returns_df['date'] = pd.to_datetime(returns_df['date'], dayfirst=True, errors='coerce')
latest_returns = returns_df.sort_values("date").groupby("Stock").tail(1)

# === Step 4: Merge Sector Info with Stock Returns ===
merged_df = pd.merge(
    latest_returns, 
    sector_df[['Ticker', 'sector']], 
    left_on='Stock', 
    right_on='Ticker', 
    how='left'
)

# === Step 5: Group by Sector and Calculate Average Return ===
sector_avg = merged_df.groupby('sector')['cumulative_return'].mean().reset_index()

#  Save Sector-Wise Performance to CSV
output_path = r"C:\Users\Nivetha s\Desktop\prject 2\3. Sector-wise Performance\sector_average_returns.csv"
sector_avg.to_csv(output_path, index=False)
print(f" Sector-wise average returns saved to: {output_path}")

# === Step 6: Plot the Sector-Wise Average Return ===
plt.figure(figsize=(12, 6))
plt.bar(sector_avg['sector'], sector_avg['cumulative_return'], color='mediumseagreen')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Average Cumulative Return")
plt.title("Sector-Wise Average Stock Performance (Past Year)")
plt.tight_layout()
plt.grid
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# === Step 1: Load All Closing Prices from Each Stock File ===
folder_path = r"C:\Users\Nivetha s\Desktop\prject 2\50 stocks"
price_data = {}

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        stock_name = filename.replace('.csv', '')
        file_path = os.path.join(folder_path, filename)

        try:
            df = pd.read_csv(file_path)
            if 'date' in df.columns and 'close' in df.columns:
                df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
                df = df.dropna(subset=['date', 'close'])
                df = df.sort_values('date')
                price_data[stock_name] = df.set_index('date')['close']
        except Exception as e:
            print(f" Error loading {filename}: {e}")

# === Step 2: Combine into One DataFrame ===
price_df = pd.DataFrame(price_data)

# === Step 3: Calculate Daily Percentage Returns ===
returns_df = price_df.pct_change().dropna()

# === Step 4: Compute Correlation Matrix ===
correlation_matrix = returns_df.corr()

# === Step 6: Save Correlation Matrix to CSV ===
output_path = r"C:\Users\Nivetha s\Desktop\prject 2\4. Stock Price Correlation\stock_price_correlation_matrix.csv"
correlation_matrix.to_csv(output_path)
print(f" Correlation matrix saved to: {output_path}")

# === Step 5: Plot the Correlation Heatmap ===
plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, linewidths=0.5, square=True)
plt.title("Stock Price Correlation Heatmap", fontsize=16)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()


```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# === Step 1: Set Folder Path ===
folder_path = r"C:\Users\Nivetha s\Desktop\prject 2\50 stocks"  # Update this to your folder path
monthly_returns_list = []

# === Step 2: Read Each Stock File and Calculate Monthly Returns ===
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        filepath = os.path.join(folder_path, filename)
        try:
            df = pd.read_csv(filepath)

            if 'date' in df.columns and 'close' in df.columns:
                df = df.dropna(subset=['close'])
                df['date'] = pd.to_datetime(df['date'], errors='coerce')
                df = df.dropna(subset=['date'])

                df['Month'] = df['date'].dt.to_period('M')
                df = df.sort_values('date')

                # Monthly return calculation
                monthly_df = df.groupby('Month').agg(
                    first_close=('close', 'first'),
                    last_close=('close', 'last')
                )
                monthly_df['monthly_return'] = ((monthly_df['last_close'] - monthly_df['first_close']) / monthly_df['first_close']) * 100
                monthly_df.reset_index(inplace=True)
                monthly_df['Stock'] = filename.replace(".csv", "")

                monthly_returns_list.append(monthly_df[['Month', 'Stock', 'monthly_return']])
        except Exception as e:
            print(f" Error processing {filename}: {e}")

# === Step 3: Combine All Monthly Returns ===
combined_returns = pd.concat(monthly_returns_list)

# === Step 4: Save Monthly Return Data to CSV ===
output_path = os.path.join(folder_path, "monthly_returns.csv")
combined_returns.to_csv(output_path, index=False)
print(f" Saved monthly returns to: {output_path}")

# === Step 5: Get Top 5 Gainers and Losers per Month ===
top_gainers = []
top_losers = []

for month in combined_returns['Month'].unique():
    month_df = combined_returns[combined_returns['Month'] == month]

    # Top 5 Gainers
    gainers = month_df.sort_values('monthly_return', ascending=False).head(5).copy()
    gainers['Rank'] = range(1, 6)
    gainers['Type'] = 'Top Gainer'
    top_gainers.append(gainers)

    # Top 5 Losers
    losers = month_df.sort_values('monthly_return').head(5).copy()
    losers['Rank'] = range(1, 6)
    losers['Type'] = 'Top Loser'
    top_losers.append(losers)

# Combine both
top_combined = pd.concat(top_gainers + top_losers)

# === Step 6: Save Gainers and Losers Data to CSV ===
top_output_path = os.path.join(folder_path, "monthly_top_5_gainers_losers.csv")
top_combined.to_csv(top_output_path, index=False)
print(f" Top 5 gainers and losers per month saved to: {top_output_path}")

# === Step 7: Visualization for Each Month ===
unique_months = top_combined['Month'].unique()
for month in unique_months:
    month_str = str(month)
    data = top_combined[top_combined['Month'] == month]

    plt.figure(figsize=(10, 6))
    colors = {'Top Gainer': 'green', 'Top Loser': 'red'}

    for category in ['Top Gainer', 'Top Loser']:
        subset = data[data['Type'] == category]
        plt.bar(subset['Stock'], subset['monthly_return'], label=category, color=colors[category])

    plt.title(f"Top 5 Gainers and Losers  {month_str}")
    plt.ylabel("Monthly Return (%)")
    plt.xlabel("Stock")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

```

```python
import pandas as pd
import mysql.connector

# Establish database connection
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Nive#1029@",
    database="stockdata"
)
cursor = connection.cursor()

# Helper function to insert data into a table
def insert_data_from_csv(csv_path, table_name, expected_columns):
    # Read entire CSV and normalize headers
    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip().str.lower()

    # Normalize expected column names too
    expected_columns_lower = [col.lower() for col in expected_columns]

    # Check if all expected columns exist
    if not all(col in df.columns for col in expected_columns_lower):
        print(f" Columns not found in {csv_path}")
        print("Expected:", expected_columns)
        print("Found:", list(df.columns))
        return

    # Use only expected columns
    df = df[expected_columns_lower]

    placeholders = ', '.join(['%s'] * len(expected_columns))
    sql = f"INSERT INTO {table_name} ({', '.join(expected_columns)}) VALUES ({placeholders})"

    for row in df.itertuples(index=False, name=None):
        cursor.execute(sql, row)

    connection.commit()
    print(f" Data inserted into '{table_name}'")

# Insert data with exact DB column names, independent of CSV formatting
insert_data_from_csv(r"C:\Users\Nivetha s\Desktop\prject 2\1. Volatility Analysis\volatility_cleaned.csv", "stock_volatility", ["Stock", "Volatility"])
insert_data_from_csv(r"C:\Users\Nivetha s\Desktop\prject 2\2. Cumulative Return Over Time\cumulative_returns_cleaned.csv", "stock_cumulative_return", ["date", "stock", "cumulative_return"])
insert_data_from_csv(r"C:\Users\Nivetha s\Desktop\prject 2\3. Sector-wise Performance\sector_returns_cleaned.csv", "sector_returns", ["sector", "cumulative_return"])
insert_data_from_csv(r"C:\Users\Nivetha s\Desktop\prject 2\4. Stock Price Correlation\stock_price_correlation_cleaned.csv", "stock_price_correlation", ["stock_name", "adanient", "adaniports", "apollohosp", "asianpaint", "axisbank", "bajaj_auto", "bajajfinsv", "bajfinance", "bel", "bhartiartl", "bpcl", "britannia", "cipla", "coalindia", "drreddy", "eichermot", "grasim", "hcltech", "hdfcbank", "hdfclife", "heromotoco", "hindalco", "hindunilvr", "icicibank", "indusindbk", "infy", "itc", "jswsteel", "kotakbank", "lt", "m_m", "maruti", "nestleind", "ntpc", "ongc", "powergrid", "reliance", "sbilife", "sbin", "shriramfin", "sunpharma", "tataconsum", "tatamotors", "tatasteel", "tcs", "techm", "titan", "trent", "ultracemco", "wipro"])
insert_data_from_csv(r"C:\Users\Nivetha s\Desktop\prject 2\5. Top 5 Gainers and Losers (Month-wise)\cleaned_monthly_top_5_gainers_losers.csv", "monthly_top_5_gainers_losers", ["month", "stock", "monthly_return", "stock_rank", "type"])

# Close the connection
cursor.close()
connection.close()
```
